```python
import pandas as pd
import numpy as np 
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
import h2o
h2o.init()
```

```python
hf=h2o.H2OFrame(df)
```

```python
from h2o.automl import H2OAutoML
x=hf.columns
y='SalePrice'
hf['SalePrice']=hf['SalePrice'].asfactor()
for r in ['Id',y]:
    x.remove(r)
    
aml = H2OAutoML(max_runtime_secs = 60)
aml.train(x = x, y = y,
          training_frame = hf)
```

```python
aml.leaderboard
```

```python
test=h2o.H2OFrame(pd.read_csv('test.csv'))
preds=aml.predict(test)
final=test.cbind(preds)
results=h2o.as_list(final)
```

```python
results.head()
```

```python
odf=results.loc[:,['Id','predict']]
odf.columns=['Id','SalePrice']
odf.to_csv('submission.csv',index=False)
```
